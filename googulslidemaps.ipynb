### Sample Keyword Dataframe Generated Using Search Payload

##### Keyword and Timeframe Payload Building
(Too many requests can incur limit error)

Significant Parameters:
* The first parameter accepts list of keywords for search data request
* The timeframe should be two dates between which is a space for data of keyword search between that timeframe or two of the same dates for date of one day
* The geo locations should be left empty (geo='') for world keyword search data or with US (geo='US') Free doms' keyword search data

```python
# Trend searched of the keyword
kw_list=['Taylor Swift']
# Timeframe of the window the keyword number of searches is searched over (2016-12-14 2017-01-25) and location (geo='' is world)
pytrends.build_payload(kw_list, cat=0, timeframe='2016-12-14 2017-01-25', geo='', gprop='')
```

##### Payload Dataframes Building

```python
# Builds dataframe of the total searches for the keyword indexed by DAYS between the timeframe dates in payload
df = pytrends.interest_over_time()

# List of dates for all days in sequential order from the timeframe in pytrends' payload
datestampsss = []
for row in df.index:
    datestampsss.append((str(row)[:10])) # Takes the first 10 characters of the date and time stamp from the dataframe index

# Builds dataframe of the total searches for the keyword in the payload timeframe (2016-12-14 2017-01-25) indexed by location (geoName)
nf = pytrends.interest_by_region()
nf['geo'] = nf.index
nf = nf[nf['geo'] != 'U.S. Outlying Islands']
nf['code'] = coco.convert(names=nf['geo'], to='ISO3')

# the first columns titled after keyword containing the total Google searches of the keyword in each region
nf
```

###Trends Dataframe Formation By Timeframe

```python
i=0 # here for exclusively as decoration

# Dataframe of keyword trends by region made for each day's date collected from the timeframe
for date in datestampsss:
    datestamp = date+" "+date # use the same dates for one day timeframe data
    pytrends.build_payload(kw_list, cat=0, timeframe=datestamp, geo='', gprop='')
    df = pytrends.interest_by_region()
    if i == 0:
      print(df.head()) # prints only the first dataframe
    i += 1

# They should output the same numbers as a new dataframe was formed but not stored for each element in the timeframes list
print(len(datestampsss))
i
```

### All Coming Together

The libraries are all integrated for a combination of Chloropleth(say that nice and slow) maps, sliders, and Goggle Trends API to form the basis of a dynamic change over time slider map of Google Trends data of searched keywords

```python
# Sample timestamps list
breeh = ['2016-12-14', '2016-12-15', '2016-12-16', '2016-12-17', '2016-12-18']

# List of map data storing geographic regional data for each day in timeframe
dmaps = []

# An individual trends payload for dates iterated in timestamps list and corresponding dataframe with geographic search data
for date in breeh:
    brih = date+" "+date
    pytrends.build_payload(kw_list, cat=0, timeframe=brih, geo='', gprop='') # makes payload with keyword for day in current iteration
    df = pytrends.interest_by_region() # dataframe of regional search data from payload
    df['geo'] = df.index # a column with the geographic location by country name labeled in the dataframe's index
    df = df[df['geo'] != 'U.S. Outlying Islands'] # the country code converter considers this conglomerate of islands fictitious
    df['code'] = coco.convert(names=df['geo'], to='ISO3') # country code conversion from the country names for chloropleth

    # building map and mapping its data for each day in timestamps list
    dmap = dict(
                type='choropleth',
                locations = df['code'], # the chloropleth uses the country code to designate location of displaying search data
                z=df[kw_list[0]].astype(float), # the upperclass way of indicating the first column of the dataframe titled after keyword
                locationmode='ISO-3', # locations are identified by this code standard
                colorscale = 'Reds',
                colorbar_title = "Poopularity", # most forget the extra 'o' when spelling this
                )

    # each data bundle for the maps will be sequentially aggregated to the maps list as the date list is iterated
    dmaps.append(dmap)


steps = []

for i in range(len(dmaps)):
    step = dict(method='restyle',
                args=['visible', [False] * len(dmaps)],
                label='Date {}'.format(breeh[i])) # label to be displayed for each step ('Date')
    step['args'][1][i] = True
    steps.append(step)



dsliders = [dict(active=0, pad={"t": 1}, steps=steps)]

layout = dict(title= "Popularity Over Time Map for " + kw_list[0],
              geo=dict(showframe=False,
                       showcoastlines=False,
                       projection={'type': 'equirectangular'}),
              sliders=dsliders)



# Combine map data and layout
fig = dict(data=dmaps, layout=layout)




# Frolic with Pleasure world swifters!
plotly.offline.iplot(fig)
```
